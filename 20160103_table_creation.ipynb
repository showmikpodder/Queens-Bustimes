```python
import sqlite3
import unicodecsv
```

```python
%load_ext sql
%sql sqlite:///2016_bustimes.db
```

```python
#creates overall database
conn = sqlite3.connect('2016_bustimes.db')
cur = conn.cursor()
```

```python
#creates stop times table for 201620160103
cur.execute("CREATE TABLE Stop_Times_201620160103(trip_id TEXT,arrival_time TEXT, departure_time TEXT, stop_id INT, stop_sequence INT, pickup_type INT, drop_off_type INT)")
```

```python
%%sql
SELECT * FROM Stop_Times_201620160103
--#check on table headers
```

```python
with open('stop_times20160103.txt', 'rb') as input_file:
    reader = unicodecsv.reader(input_file, delimiter = ',')
    data = [row for row in reader]
cur.executemany("INSERT INTO Stop_Times_201620160103 (trip_id, arrival_time, departure_time, stop_id, stop_sequence, pickup_type, drop_off_type) VALUES (?, ?, ?, ?, ?, ?, ?);", data)
#reading indata from text file into the stop times 201620160103 table
```

```python
conn.commit()
```

```python
%%sql
SELECT * FROM Stop_Times_201620160103
LIMIT 5
--#check on table 
```
