```python
%%sql
DROP TABLE IF EXISTS overall_Stop_Times_2016;
CREATE TABLE IF NOT EXISTS overall_Stop_Times_2016 AS
    SELECT
        *
    FROM
        Stop_Times_20160103
    UNION ALL
    SELECT
        *
    FROM
        Stop_Times_20160403  
    UNION ALL
    SELECT
        *
    FROM
        Stop_Times_20160703
--#joins all three stop_times tables into one
```

```python
%%sql
SELECT COUNT(DISTINCT(trip_id)) FROM overall_Stop_Times_2016
--#all stop_times in 2016 
```

```python
%%sql
DROP TABLE IF EXISTS overall_Trips_2016;
CREATE TABLE IF NOT EXISTS overall_Trips_2016 AS
    SELECT
        *
    FROM
        Trips_20160103
    UNION ALL
    SELECT
        *
    FROM
        Trips_20160403  
    UNION ALL
    SELECT
        *
    FROM
        Trips_20160703
--#joins all three trips tables into one
```

```python
%%sql
SELECT COUNT(trip_id) FROM overall_Trips_2016
--#all trips in 2016 
```

```python
%%sql
DROP TABLE IF EXISTS combined_Trips_Stop_Times_2016;
CREATE TABLE IF NOT EXISTS combined_Trips_Stop_Times_2016 AS
    SELECT
        overall_Trips_2016.trip_id, 
        overall_Trips_2016.route_id,
        overall_Trips_2016.direction_id,
        overall_Stop_Times_2016.arrival_time,
        overall_Stop_Times_2016.departure_time,
        overall_Stop_Times_2016.stop_id,
        overall_Stop_Times_2016.stop_sequence
    FROM 
        overall_Trips_2016
    INNER JOIN
        overall_Stop_Times_2016
    ON
        overall_Trips_2016.trip_id = overall_Stop_Times_2016.trip_id;
```

```python
%%sql
SELECT COUNT(DISTINCT(trip_id)) FROM combined_Trips_Stop_Times_2016;
SELECT * FROM combined_Trips_Stop_Times_2016 LIMIT 5464, 2
```

```python
%%sql
DROP TABLE IF EXISTS q111_weekday_2016trips;
CREATE TABLE IF NOT EXISTS q111_weekday_2016trips AS
   SELECT 
        *
    FROM 
        combined_Trips_Stop_Times_2016 
    WHERE 
        route_id is 'Q111' and trip_id LIKE '%Weekday%'
--#weekday q111 trips
```

```python
%%sql
DROP TABLE IF EXISTS q111_weekday_2016trips_stops;
CREATE TABLE IF NOT EXISTS q111_weekday_2016trips_stops AS
    SELECT
        q111_weekday_2016trips.trip_id, 
        q111_weekday_2016trips.route_id,
        q111_weekday_2016trips.direction_id,
        q111_weekday_2016trips.arrival_time,
        q111_weekday_2016trips.departure_time,
        q111_weekday_2016trips.stop_id,
        q111_weekday_2016trips.stop_sequence,
        Stops_20160103.stop_name,
        Stops_20160103.stop_desc
        
    FROM 
        q111_weekday_2016trips
    INNER JOIN
        Stops_20160103
    ON
        q111_weekday_2016trips.stop_id = Stops_20160103.stop_id;
--#combining in information from the stops_desc table
```

```python
%%sql
SELECT
    *
FROM
    q111_weekday_2016trips_stops
LIMIT 200
```

```python
%%sql
DROP TABLE IF EXISTS q111_weekday_2016_FranLew_147Dr_departure;
CREATE TABLE IF NOT EXISTS q111_weekday_2016_FranLew_147Dr_departure AS
    SELECT
        *
    FROM
        q111_weekday_2016trips_stops
    WHERE
        stop_name is 'FRANCIS LEWIS BL /147 DR';
SELECT * FROM q111_weekday_2016_FranLew_147Dr_departure limit 2
```

```python
%%sql
DROP TABLE IF EXISTS q111_weekday_2016_hillsde_parsn_arrival;
CREATE TABLE IF NOT EXISTS q111_weekday_2016_hillsde_parsn_arrival AS
    SELECT
        *
    FROM
        q111_weekday_2016trips_stops
    WHERE
        stop_name is 'HILLSIDE AV/PARSONS BL';
SELECT * FROM q111_weekday_2016_hillsde_parsn_arrival limit 2
```

```python
%%sql
DROP TABLE IF EXISTS q111_weekday_2016_full_trip_times_temp;
CREATE TABLE IF NOT EXISTS q111_weekday_2016_full_trip_times_temp AS
    SELECT
        q111_weekday_2016_FranLew_147Dr_departure.trip_id, 
        q111_weekday_2016_FranLew_147Dr_departure.route_id,
        q111_weekday_2016_FranLew_147Dr_departure.direction_id,
        q111_weekday_2016_FranLew_147Dr_departure.departure_time,
        q111_weekday_2016_FranLew_147Dr_departure.stop_sequence,
        q111_weekday_2016_FranLew_147Dr_departure.stop_name as first_stop,
        q111_weekday_2016_hillsde_parsn_arrival.arrival_time,
        q111_weekday_2016_hillsde_parsn_arrival.stop_name as last_stop    
        
    FROM 
        q111_weekday_2016_FranLew_147Dr_departure
    INNER JOIN
        q111_weekday_2016_hillsde_parsn_arrival
    ON
        q111_weekday_2016_FranLew_147Dr_departure.trip_id = q111_weekday_2016_hillsde_parsn_arrival.trip_id;
SELECT * FROM q111_weekday_2016_full_trip_times_temp
```

```python
%%sql
DROP TABLE IF EXISTS q111_weekday_2016_full_trip_times_temp2;
CREATE TABLE IF NOT EXISTS q111_weekday_2016_full_trip_times_temp2 AS
SELECT * FROM q111_weekday_2016_full_trip_times_temp WHERE stop_sequence is 1
```

```python
%%sql
DROP TABLE IF EXISTS q111_weekday_2016_full_trip_times_final;
CREATE TABLE IF NOT EXISTS q111_weekday_2016_full_trip_times_final AS
    SELECT
        trip_id,
        route_id,
        first_stop,
        departure_time,
        last_stop,
        arrival_time
    FROM
        q111_weekday_2016_full_trip_times_temp2
        
```

```python
%%sql
SELECT * FROM
q111_weekday_2016_full_trip_times_final
ORDER BY departure_time
```

```python
%%sql
DROP TABLE IF EXISTS q111_weekday_2016_simplify_temp;
CREATE TABLE IF NOT EXISTS q111_weekday_2016_simplify_temp AS
    SELECT
        trip_id,
        route_id,
        first_stop,
        departure_time,
        last_stop,
        arrival_time,
        ((strftime('%s', arrival_time) - strftime('%s', departure_time))/ 60) as trip_time
    FROM
        q111_weekday_2016_full_trip_times_final;
SELECT * FROM q111_weekday_2016_simplify_temp
```

```python
%%sql
DROP TABLE IF EXISTS q111_weekday_2016_simplify_final;
CREATE TABLE IF NOT EXISTS q111_weekday_2016_simplify_final AS
    SELECT
        trip_id,
        departure_time,
        trip_time,
        trip_time-40 as score
    FROM
        q111_weekday_2016_simplify_temp;
SELECT * FROM q111_weekday_2016_simplify_final
        
```
