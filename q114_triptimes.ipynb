```python
import sqlite3
import unicodecsv
import pandas as pd
import numpy as np
import datetime
```

```python
%%sql
DROP TABLE IF EXISTS q85_weekday_2016trips;
DROP TABLE IF EXISTS q114_weekday_2016trips;
CREATE TABLE IF NOT EXISTS q114_weekday_2016trips AS
   SELECT 
        *
    FROM 
        combined_Trips_Stop_Times_2016 
    WHERE 
        route_id is 'Q114' and trip_id LIKE '%Weekday%'
--#weekday q114 trips
```

```python
%%sql
DROP TABLE IF EXISTS q114_weekday_2016trips_stops;
CREATE TABLE IF NOT EXISTS q114_weekday_2016trips_stops AS
    SELECT
        q114_weekday_2016trips.trip_id, 
        q114_weekday_2016trips.route_id,
        q114_weekday_2016trips.direction_id,
        q114_weekday_2016trips.arrival_time,
        q114_weekday_2016trips.departure_time,
        q114_weekday_2016trips.stop_id,
        q114_weekday_2016trips.stop_sequence,
        Stops_20160103.stop_name,
        Stops_20160103.stop_desc
        
    FROM 
        q114_weekday_2016trips
    INNER JOIN
        Stops_20160103
    ON
        q114_weekday_2016trips.stop_id = Stops_20160103.stop_id;
--#combining in information from the stops_desc table
```

```python
%%sql
SELECT * FROM q114_weekday_2016trips_stops
LIMIT 50
```

```python
%%sql
DROP TABLE IF EXISTS q114_weekday_2016_hillsde_parsn_arrival;
CREATE TABLE IF NOT EXISTS q114_weekday_2016_hillsde_parsn_arrival AS
    SELECT
        *
    FROM
        q114_weekday_2016trips_stops
    WHERE
        stop_name is 'HILLSIDE AV/PARSONS BL';
SELECT * FROM q114_weekday_2016_hillsde_parsn_arrival limit 10
```

```python
%%sql
DROP TABLE IF EXISTS q114_weekday_2016_crest_Rd_departure;
CREATE TABLE IF NOT EXISTS q114_weekday_2016_crest_Rd_departure AS
    SELECT
        *
    FROM
        q114_weekday_2016trips_stops
    WHERE
        stop_name is 'SEAGIRT BL /CREST RD';
SELECT * FROM q114_weekday_2016_crest_Rd_departure limit 2
```

```python
%%sql
DROP TABLE IF EXISTS q114_weekday_2016_full_trip_times_temp;
CREATE TABLE IF NOT EXISTS q114_weekday_2016_full_trip_times_temp AS
    SELECT
        q114_weekday_2016_crest_Rd_departure.trip_id, 
        q114_weekday_2016_crest_Rd_departure.route_id,
        q114_weekday_2016_crest_Rd_departure.direction_id,
        q114_weekday_2016_crest_Rd_departure.departure_time,
        q114_weekday_2016_crest_Rd_departure.stop_sequence,
        q114_weekday_2016_crest_Rd_departure.stop_name as first_stop,
        q114_weekday_2016_hillsde_parsn_arrival.arrival_time,
        q114_weekday_2016_hillsde_parsn_arrival.stop_name as last_stop    
        
    FROM 
        q114_weekday_2016_crest_Rd_departure
    INNER JOIN
        q114_weekday_2016_hillsde_parsn_arrival
    ON
        q114_weekday_2016_crest_Rd_departure.trip_id = q114_weekday_2016_hillsde_parsn_arrival.trip_id;
SELECT * FROM q114_weekday_2016_full_trip_times_temp
```

```python
%%sql
DROP TABLE IF EXISTS q114_weekday_2016_full_trip_times_temp2;
CREATE TABLE IF NOT EXISTS q114_weekday_2016_full_trip_times_temp2 AS
SELECT * FROM q114_weekday_2016_full_trip_times_temp WHERE stop_sequence is 1
```

```python
%%sql
DROP TABLE IF EXISTS q114_weekday_2016_full_trip_times_final;
CREATE TABLE IF NOT EXISTS q114_weekday_2016_full_trip_times_final AS
    SELECT
        trip_id,
        route_id,
        first_stop,
        departure_time,
        last_stop,
        arrival_time
    FROM
        q114_weekday_2016_full_trip_times_temp2
```

```python
%%sql
SELECT * FROM
q114_weekday_2016_full_trip_times_final
ORDER BY departure_time
```

```python
%%sql
DROP TABLE IF EXISTS q114_weekday_2016_simplify_temp;
CREATE TABLE IF NOT EXISTS q114_weekday_2016_simplify_temp AS
    SELECT
        trip_id,
        route_id,
        first_stop,
        departure_time,
        last_stop,
        arrival_time,
        ((strftime('%s', arrival_time) - strftime('%s', departure_time))/ 60) as trip_time
    FROM
        q114_weekday_2016_full_trip_times_final;
SELECT * FROM q114_weekday_2016_simplify_temp
```

```python
%%sql
DROP TABLE IF EXISTS q114_weekday_2016_simplify_final;
CREATE TABLE IF NOT EXISTS q114_weekday_2016_simplify_final AS
    SELECT
        trip_id,
        departure_time,
        trip_time,
        trip_time-59 as score
    FROM
        q114_weekday_2016_simplify_temp;
SELECT * FROM q114_weekday_2016_simplify_final
```

```python
%%sql
SELECT * FROM
q114_weekday_2016_simplify_final
ORDER BY departure_time
```

```python
%%sql
DROP TABLE IF EXISTS avg_score_q114;
CREATE TABLE IF NOT EXISTS avg_score_q114 AS
    SELECT 
        DISTINCT(departure_time) as departure_time, AVG(score) as average_score
    FROM
        q114_weekday_2016_simplify_final
    GROUP BY
        departure_time
```

```python
%%sql
SELECT * FROM avg_score_q114
```
