```python
q114_bus_schedule = pd.read_csv('q114_bus.csv')
q114_bus_schedule.columns = ["Seagrit Bl/Crest Rd", " Mott Av/ Central Av", " G Brwr Bl/Farmers Bl", " G Brwr Bl/115 Av", "Parsons Bl/88 Av"]
q114_bus_schedule
```

```python
q114_bus_schedule['Seagrit Bl/Crest Rd'] = q114_bus_schedule['Seagrit Bl/Crest Rd'].replace('None', np.nan)
```

```python
q114_bus_schedule["Parsons Bl/88 Av"] = pd.to_datetime(q114_bus_schedule["Parsons Bl/88 Av"]).apply(lambda x: x.strftime(r'%H:%M') if pd.notnull(x) else '')
q114_bus_schedule['Seagrit Bl/Crest Rd'] = pd.to_datetime(q114_bus_schedule['Seagrit Bl/Crest Rd']).apply(lambda x: x.strftime(r'%H:%M') if pd.notnull(x) else '')
```

```python
q114_bus_schedule["Parsons Bl/88 Av"] = pd.to_datetime(q114_bus_schedule["Parsons Bl/88 Av"])
q114_bus_schedule['Seagrit Bl/Crest Rd'] = pd.to_datetime(q114_bus_schedule['Seagrit Bl/Crest Rd'])
```

```python
q114_bus_schedule['Total Trip Time'] = q114_bus_schedule["Parsons Bl/88 Av"] - q114_bus_schedule['Seagrit Bl/Crest Rd']
```
